```python
#counting how many NaN we have
count2018 = 0
count2019 = 0

for i in df['International  tourist  arrivals  (2018)']:
    if  i == '-':
        count2018 = count2018 + 1

for i in df['International  tourist  arrivals  (2019)']:
    if  i == '–':
        count2019 = count2019 + 1

#Pay attention to '-' is not the same simbol as '–'
print(count2018, count2019)
```

```python
#changing those simbols with 0
df.loc[df['International  tourist  arrivals  (2018)'] == '-' , 'International  tourist  arrivals  (2018)'] = 0
df.loc[df['International  tourist  arrivals  (2019)'] == '–' , 'International  tourist  arrivals  (2019)'] = 0
```

```python
#I will keep the first 4 string values, fill with 0 the resultant NaNs and transform the result in float
df['International  tourist  arrivals  (2018)'] = df['International  tourist  arrivals  (2018)'].str[:4].fillna(0).astype(float)
df['International  tourist  arrivals  (2019)'] = df['International  tourist  arrivals  (2019)'].str[:4].fillna(0).astype(float)
```

```python
#transforming the values into millions
df['International  tourist  arrivals  (2018)'] = df['International  tourist  arrivals  (2018)'] * 1000000
df['International  tourist  arrivals  (2019)'] = df['International  tourist  arrivals  (2019)'] * 1000000
```

```python
df['International  tourist  arrivals  (2018)'] = df['International  tourist  arrivals  (2018)'].astype(int)
df['International  tourist  arrivals  (2019)'] = df['International  tourist  arrivals  (2019)'].astype(int)
```

```python
pd.pivot_table(df,index=['Region','Destination'],
 values=['International  tourist  arrivals  (2018)','International  tourist  arrivals  (2019)'])
```
